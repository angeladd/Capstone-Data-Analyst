```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Define the URL of the dataset
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/VYPrOu0Vs3I0hKLLjiPGrA/survey-data-with-duplicate.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example: Top 10 programming languages
top_languages = df['LanguageHaveWorkedWith'].value_counts().head(10)

# Set up the figure
plt.figure(figsize=(10, 6))

# Bar positions and labels
languages = top_languages.index
counts = top_languages.values
colors = plt.cm.tab10.colors  # 10 distinct colors

# Plot each bar individually with a different color
plt.bar(languages, counts, color=colors[:len(languages)])

# Add titles and labels
plt.title('Top 10 Programming Languages Used (Current Year)', fontsize=14)
plt.xlabel('Programming Language', fontsize=12)
plt.ylabel('Number of Respondents', fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python

# Example: language usage over two years (mock data)
data = {
    'Language': ['Python', 'JavaScript', 'Java', 'C#', 'C++'],
    '2024': [320, 280, 200, 150, 130]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Assume simple growth rate of 10% to predict 2025
df['2025'] = (df['2024'] * 1.10).astype(int)

# Plot side-by-side bar chart
x = range(len(df['Language']))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar([i - width/2 for i in x], df['2024'], width=width, label='2024', color='skyblue')
plt.bar([i + width/2 for i in x], df['2025'], width=width, label='2025 (Predicted)', color='orange')

plt.xticks(ticks=x, labels=df['Language'])
plt.title('Programming Language Usage: 2024 vs 2025 Prediction')
plt.xlabel('Language')
plt.ylabel('Number of Users')
plt.legend()
plt.tight_layout()
plt.show()
```

```python

# Sample data (replace this with your actual top 10 databases data)
data = {
    'Database': ['MySQL', 'PostgreSQL', 'SQLite', 'MongoDB', 'Oracle', 
                 'SQL Server', 'Firebase', 'Redis', 'Elasticsearch', 'MariaDB'],
    'Count': [1200, 1150, 980, 870, 750, 700, 650, 600, 570, 540]
}

df = pd.DataFrame(data)

# Generate a unique color for each bar
colors = plt.cm.tab10.colors  # default 10 distinct colors
# If more than 10, use another colormap or repeat
colors = colors[:len(df)]

# Plotting
plt.figure(figsize=(12, 6))
bars = plt.bar(df['Database'], df['Count'], color=colors)

# Add value labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 20, int(yval), ha='center', va='bottom')

plt.title('Top 10 Databases (Current Year)', fontsize=16)
plt.xlabel('Database')
plt.ylabel('Usage Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# Current year database usage (replace with your actual data)
current_data = {
    'Database': ['MySQL', 'PostgreSQL', 'SQLite', 'MongoDB', 'Oracle',
                 'SQL Server', 'Firebase', 'Redis', 'Elasticsearch', 'MariaDB'],
    'Count': [1200, 1150, 980, 870, 750, 700, 650, 600, 570, 540]
}

# Assume a 10% growth for projection
growth_rate = 0.10
df_current = pd.DataFrame(current_data)
df_current['Projected_Next_Year'] = (df_current['Count'] * (1 + growth_rate)).round(0).astype(int)

# Sort by projected usage
df_projected = df_current.sort_values(by='Projected_Next_Year', ascending=False)

# Assign different colors
colors = plt.cm.Set3.colors[:len(df_projected)]

# Plotting
plt.figure(figsize=(12, 6))
bars = plt.bar(df_projected['Database'], df_projected['Projected_Next_Year'], color=colors)

# Add value labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 20, int(yval), ha='center', va='bottom')

plt.title("Projected Top 10 Databases for Next Year", fontsize=16)
plt.xlabel("Database")
plt.ylabel("Projected Usage Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample structure (replace with your actual dataset)
# Assume languages are stored like "Python; JavaScript; SQL"
df = pd.DataFrame({
    'Gender': ['Woman', 'Woman', 'Man', 'Woman'],
    'LanguageHaveWorkedWith': [
        'Python;SQL',
        'JavaScript;HTML/CSS',
        'Python;C++',
        'Python;JavaScript;SQL'
    ]
})

# Step 1: Filter only women
women_df = df[df['Gender'] == 'Woman']

# Step 2: Split and flatten the language list
all_langs = women_df['LanguageHaveWorkedWith'].dropna().str.split(';').explode()

# Step 3: Count frequency
lang_counts = all_langs.value_counts()

# Step 4: Plot pie chart
plt.figure(figsize=(8, 8))
plt.pie(lang_counts, labels=lang_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Set3.colors)
plt.title("Preferred Programming Languages Used by Women")
plt.axis('equal')  # Keep pie chart circular
plt.tight_layout()
plt.show()
```

```python


# Data for the languages and average annual salaries
data = {
    'Language': ['Python', 'Java', 'R', 'Javascript', 'Swift', 'C++', 'C#', 'PHP', 'SQL', 'Go'],
    'Average Annual Salary': ['$114,383.00', '$101,013.00', '$92,037.00', '$110,981.00', '$130,801.00', 
                             '$113,865.00', '$88,726.00', '$84,727.00', '$84,793.00', '$94,082.00']
}

# Create DataFrame
df = pd.DataFrame(data)

# Remove dollar sign and commas, and convert to float for sorting
df['Average Annual Salary'] = df['Average Annual Salary'].replace({'\$': '', ',': ''}, regex=True).astype(float)

# Step 1: Sort the DataFrame by salary in ascending order
df_sorted = df.sort_values(by='Average Annual Salary', ascending=True)

# Step 2: Create a color palette
colors = plt.cm.get_cmap('tab10', len(df_sorted))  # Use a color palette with enough colors for the bars

# Step 3: Plot the bar chart with different colors for each bar
plt.figure(figsize=(10, 6))
bars = plt.bar(df_sorted['Language'], df_sorted['Average Annual Salary'], color=colors(np.arange(len(df_sorted))), edgecolor='black')

# Adding labels and title
plt.title('Programming Languages and Average Annual Salary', fontsize=14)
plt.xlabel('Programming Language', fontsize=12)
plt.ylabel('Average Annual Salary (in $)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample data for job postings by language
data = {
    'Language': ['Python', 'Java', 'JavaScript', 'C++', 'SQL', 'PHP', 'Go', 'R', 'Swift', 'C#'],
    'Job Postings': [12000, 9500, 10500, 8000, 13500, 6000, 4500, 7000, 5000, 8500]
}

# Create DataFrame
df = pd.DataFrame(data)

# Step 1: Sort the DataFrame by job postings in ascending order
df_sorted = df.sort_values(by='Job Postings', ascending=True)

# Step 2: Create a color palette with distinct colors for each bar
colors = plt.cm.get_cmap('tab10', len(df_sorted))  # Using 'tab10' color map

# Step 3: Plot the bar chart with different colors for each bar
plt.figure(figsize=(10, 6))
bars = plt.bar(df_sorted['Language'], df_sorted['Job Postings'], color=colors(np.arange(len(df_sorted))), edgecolor='black')

# Adding labels and title
plt.title('Job Postings by Programming Language (Ascending Order)', fontsize=14)
plt.xlabel('Programming Language', fontsize=12)
plt.ylabel('Number of Job Postings', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```
